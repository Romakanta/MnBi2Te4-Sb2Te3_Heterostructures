## FFNN on magnetic moment without classification

```python
# import absl.logging
# absl.logging.set_verbosity(absl.logging.ERROR)

import tensorflow as tf
from tensorflow import keras
import tensorflow_addons as tfa
import pickle

from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam, SGD
from sklearn import metrics
from tensorflow.keras.models import load_model

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
print(tf.__version__)
```

```python
df = pd.read_excel('STEP5_pearson_corr_less_than_98%_from_plots_reasonable.xlsx')
```

```python
Xy = df.drop(['material','spin_state','formation_energy','Ef_per_cell','bandgap','mag_mom_table','sitewise_axiswise_tot_mu','sitewise_mu_mag','axiswise_tot_mu',
   'total_mag','MBT','BT','MBT_BT_combined','chemical_composition'],axis=1)
```

```python
train_val_dataset = Xy.sample(frac = 0.8, random_state = 42)
train_dataset = train_val_dataset.sample(frac = 0.75, random_state = 42)
val_dataset = train_val_dataset.drop(train_dataset.index)
test_dataset = Xy.drop(train_val_dataset.index)

train_data = np.copy(train_dataset)
val_data = np.copy(val_dataset)
test_data = np.copy(test_dataset)

train_data = pd.DataFrame(train_data)
val_data = pd.DataFrame(val_data)
test_data = pd.DataFrame(test_data)

X_train = train_data.drop([0],axis='columns')
X_val = val_data.drop([0],axis='columns')
X_test = test_data.drop([0],axis='columns')

y_train = train_data.drop(train_data.iloc[:,1:],axis=1)
y_val = val_data.drop(val_data.iloc[:,1:],axis=1)
y_test = test_data.drop(test_data.iloc[:,1:],axis=1)
```

```python
def build_and_compile_model(norm):
    model=tf.keras.Sequential([
        norm,
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(1)
    ])
    
    model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
                 loss='mean_absolute_error')
    
    return model
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(np.array(X_train))

dnn_mu = build_and_compile_model(normalizer)
history_mu=dnn_mu.fit(x=X_train, y=y_train, validation_split=0.25,verbose=0, epochs=500)
```

```python
r2_train = metrics.r2_score(y_train, dnn_mu.predict(X_train))
r2_val = metrics.r2_score(y_val, dnn_mu.predict(X_val))
r2_test = metrics.r2_score(y_test, dnn_mu.predict(X_test))
print(r2_train,r2_val,r2_test)
```

```python
plt.scatter(y_train, dnn_mu.predict(X_train), color='green', s=100, marker='o', label='Train($R^2$ = '+str(f'{r2_train:.3f}')+')')
plt.scatter(y_test, dnn_mu.predict(X_test), color='red',s=100, marker='s', label='Test($R^2$ = '+str(f'{r2_test:.3f}')+')')
plt.plot([0,0.5], [0,0.5],linewidth=2.5, color='gray', linestyle='dashed')
plt.xlabel('DFT Mu'), plt.ylabel('Predicted Mu') 
plt.legend()
plt.show()
```

```python
## Running a model for 200 times
```

```python
def build_and_compile_model(norm):
    model=tf.keras.Sequential([
        norm,
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(32, activation='relu'),
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(1)
    ])
    
    model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
                 loss='mean_absolute_error')
    
    return model

dnn_mu_list = []
train_mu_r2 = []
val_mu_r2 = []
test_mu_r2 = []

for i in range(50):
    
    train_val_dataset = Xy.sample(frac = 0.8, random_state = i)
    train_dataset = train_val_dataset.sample(frac = 0.75, random_state = i)
    val_dataset = train_val_dataset.drop(train_dataset.index)
    test_dataset = Xy.drop(train_val_dataset.index)
    
    train_data = np.copy(train_dataset)
    val_data = np.copy(val_dataset)
    test_data = np.copy(test_dataset)
    
    train_data = pd.DataFrame(train_data)
    val_data = pd.DataFrame(val_data)
    test_data = pd.DataFrame(test_data)
    
    X_train = train_data.drop([0],axis='columns')
    X_val = val_data.drop([0],axis='columns')
    X_test = test_data.drop([0],axis='columns')
    
    y_train = train_data.drop(train_data.iloc[:,1:],axis=1)
    y_val = val_data.drop(val_data.iloc[:,1:],axis=1)
    y_test = test_data.drop(test_data.iloc[:,1:],axis=1)

    X_train.to_csv('Results_Mu_new/train_test_data_200/X_train_'+str(i)+'.csv',index=False)
    X_val.to_csv('Results_Mu_new/train_test_data_200/X_val_'+str(i)+'.csv',index=False)
    X_test.to_csv('Results_Mu_new/train_test_data_200/X_test_'+str(i)+'.csv',index=False)
    y_train.to_csv('Results_Mu_new/train_test_data_200/y_train_'+str(i)+'.csv',index=False)
    y_val.to_csv('Results_Mu_new/train_test_data_200/y_val_'+str(i)+'.csv',index=False)
    y_test.to_csv('Results_Mu_new/train_test_data_200/y_test_'+str(i)+'.csv',index=False)
    
    normalizer = tf.keras.layers.Normalization(axis=-1)

    normalizer.adapt(np.array(X_train))
    
    dnn_mu = build_and_compile_model(normalizer)

    history_mu=dnn_mu.fit(x=X_train, y=y_train, validation_split=0.25,verbose=0, epochs=1000)
    
    dnn_mu_list.append(dnn_mu)
    dnn_mu.save(f'Results_Mu_new/dnn_mu_{i}')#'.h5',save_format = 'tf')
    
    with open(f'Results_Mu_new/dnn_mu_{i}.pkl', 'wb') as f:
        pickle.dump(dnn_mu.history, f)

    train_mu_r2.append(metrics.r2_score(y_train, dnn_mu.predict(X_train)))
    val_mu_r2.append(metrics.r2_score(y_val, dnn_mu.predict(X_val)))
    test_mu_r2.append(metrics.r2_score(y_test, dnn_mu.predict(X_test)))
    print('Loop number = ', i+1, 'is finished.')
```
