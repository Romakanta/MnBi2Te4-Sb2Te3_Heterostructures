```python
df[['MBT', 'BT']] = df['material'].str.split('_', expand=True)
```

```python
df
```

```python
mbt=df['MBT']
bt=df['BT']
material_dense_form=[str(mbt)+str(bt) for mbt, bt in zip(mbt,bt)]
```

```python
material_dense_form
```

```python
df['MBT_BT_combined'] = [i for i in material_dense_form]
```

```python
def split_material(material_str):
    elements = re.findall(r"([A-Z][a-z]*)(\d+)", material_str)
    
    expanded_elements = []
    for element, count in elements:
        expanded_elements.extend([element] * int(count)) 
    return expanded_elements

df['chemical_composition'] = df['MBT_BT_combined'].apply(split_material)
```
