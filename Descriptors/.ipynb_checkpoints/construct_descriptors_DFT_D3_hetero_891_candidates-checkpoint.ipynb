```python
def split_material(material_str):
    elements = re.findall(r"([A-Z][a-z]*)(\d+)", material_str)
    
    expanded_elements = []
    for element, count in elements:
        expanded_elements.extend([element] * int(count)) 
    return expanded_elements

df['chemical_composition'] = df['MBT_BT_combined'].apply(split_material)
```

```python
template_MBT_BT_BX_no_split = ['A1', 'A2', 'B1', 'X1', 'B2', 'X2']

element_A = ['Mn', 'Sc', 'Ti', 'V', 'Cr', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Y', 'Zr', 'Nb', 'Mo', 'Tc', 'Ru', 'Rh', 'Pd', 'Ag', 'Cd', 'Hf',
      'Ta', 'W', 'Re', 'Os', 'Ir', 'Pt', 'Au', 'Hg']
      
element_B = ['Sn', 'Sb', 'Pb', 'Bi', 'Si']

element_X = ['S', 'Se', 'Te']

element_ABX_combined = element_A + element_B + element_X
```

```python
ion_energy = []
vdw_radius = []
dipole_pol = []
no_electrons = []
atomic_volume = []
atomic_radius = []
covalent_radius = []
valence_electrons = []
electron_affinity = []
electronegativity = []
unpaired_electrons = []

## Newly added
period = []
density = []
atomic_mass = []
specific_heat = []
boiling_point = []
melting_point = []
metallic_radius = []
proton_affinity = []
mendeleev_number = []
lattice_constant = []
heat_of_formation = []

for i in element_ABX_combined:
    ion_energy.append(element(i).ionenergies[1]) 
    vdw_radius.append(element(i).vdw_radius) 
    dipole_pol.append(element(i).dipole_polarizability)
    no_electrons.append(element(i).atomic_number) 
    atomic_volume.append(element(i).atomic_volume) 
    atomic_radius.append(element(i).atomic_radius) 
    covalent_radius.append(element(i).covalent_radius) 
    valence_electrons.append(element(i).nvalence()) 
    electron_affinity.append(element(i).electron_affinity) 
    electronegativity.append(element(i).en_pauling) 
    unpaired_electrons.append(element(i).ec.unpaired_electrons()) 

    period.append(element(i).period) 
    density.append(element(i).density) 
    atomic_mass.append(element(i).mass) 
    specific_heat.append(element(i).specific_heat) 
    boiling_point.append(element(i).boiling_point) 
    melting_point.append(element(i).melting_point) 
    metallic_radius.append(element(i).metallic_radius) 
    proton_affinity.append(element(i).proton_affinity) 
    mendeleev_number.append(element(i).mendeleev_number) 
    lattice_constant.append(element(i).lattice_constant) 
    heat_of_formation.append(element(i).heat_of_formation)
```
