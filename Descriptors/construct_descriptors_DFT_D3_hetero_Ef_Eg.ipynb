```python
def split_material(material_str):
    elements = re.findall(r"([A-Z][a-z]*)(\d+)", material_str)
    
    expanded_elements = []
    for element, count in elements:
        expanded_elements.extend([element] * int(count)) 
    return expanded_elements

df['chemical_composition'] = df['MBT_BT_combined'].apply(split_material)
```
